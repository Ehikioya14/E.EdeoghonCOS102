# Program to calculate Simple interest, Compound interest , Annuity plan
def simple_interest():
    print("\nSimple Interest")
    P = float(input("Enter the Principal (P) : "))
    R = float(input("Enter the Rate (%): "))
    T = float(input("Enter the Time in years (T): "))

    # define the simple interest formula
    S_I = P*(1 + (R/100)*T)
    print(f"The simple interest is: {S_I}")

def compound_interest():
    print("\nCompound interest")
    P = float(input("Enter the Principal (P) : "))
    R = float(input("Enter the Rate (%): "))
    n = int(input("Number of times interest is compounded (n): "))
    T = float(input("Enter the Time in years (T): "))

    # define the compound interest formula
    C_I = P*(1 + (R/n))**(n*T)
    print(f"The compound interest is: {C_I}")

def annuity_plan():
    print("\nAnnuity Plan")
    P = float(input("Enter the Principal (P) : "))
    R = float(input("Enter the Rate (%): "))
    M = float(input("Enter the maturity value (M):"))          
    n = int(input("Number of times interest is compounded (n): "))
    T = float(input("Enter the Time in years (T): "))

    # define the annuity plan formula
    A = (P*M*T)*(((1 +(R/n))**(n*T)) - 1) / (R/n)
    print(f"The Annuity plan is: {A}")

while True:
    print("\n----Calculator for Simple Interest,Compound Interest,Annuity plan------")
    print("Pick betwwen 1 - 4 ")
    print("1. Simple Interest")
    print("2. Compound Interest")
    print("3. Annuity plan")
    print("4. Exit")

    choice = int(input())
    if choice == 1:
        simple_interest()
    elif choice == 2:
        compound_interest()
    elif choice == 3:
        annuity_plan()
    elif choice == 4:
        print("Exiting........")
        break
    else:
        print("Error, invalid input")
