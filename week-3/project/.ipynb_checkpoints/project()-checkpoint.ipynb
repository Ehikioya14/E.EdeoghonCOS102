```python

import tkinter as tk
from tkinter import messagebox
import pandas as pd


# Load the dataset 
try:
    data = pd.read_excel(r"C:\Users\JOHN EDEOGHON\OneDrive\Documents\E.EdeoghonCOS102\week-3\project\GIG-logistics.xlsx")
except FileNotFoundError:
    messagebox.showerror("Error", "CSV file not found!")
    exit()

# Function to check employee
def check_employee():
    name = name_entry.get()
    department = dept_entry.get()

    # Check if inputs are empty
    if name == "" or department == "":
        messagebox.showwarning("Input Error", "Please enter both name and department")
        return

    # Convert to lowercase for easy matching
    name_lower = name.lower()
    dept_lower = department.lower()

    # Search for employee
    found = data[
        (data["Name"].str.lower() == name_lower) &
        (data["Department"].str.lower() == dept_lower)
    ]

    if not found.empty:
        # Employee exists
        result_text.set(f"Welcome {name}!\n")

        # Get other employees in same department
        dept_members = data[data["Department"].str.lower() == dept_lower]

        members_list = "Other members in this department:\n"

        for i in dept_members["Name"]:
            members_list += f"- {i}\n"

        result_text.set(result_text.get() + members_list)

    else:
        messagebox.showinfo("Result", "Employee does not exist")


# Create main window
root = tk.Tk()
root.title("GIG Logistics Employee Verification")
root.geometry("400x400")


# Create Labels and Entry fields
tk.Label(root, text="Enter Name").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack(pady=5)

tk.Label(root, text="Enter Department").pack(pady=5)
dept_entry = tk.Entry(root)
dept_entry.pack(pady=5)


# Create Button
tk.Button(root, text="Check Employee", command=check_employee).pack(pady=10)


# Output display
result_text = tk.StringVar()
result_label = tk.Label(root, textvariable=result_text, justify="left")
result_label.pack(pady=10)


# Run the app
root.mainloop()
```
