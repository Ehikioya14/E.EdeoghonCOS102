```python
import tkinter as tk
import random

class Employee:
    def __init__(self):
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso",
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]

        self.tasks = [
            "Loading", "Transporting", "Reviewing Orders",
            "Customer Service", "Delivering Items"
        ]

    def check_employee(self, name):
        return name.strip().title() in [emp.strip().title() for emp in self.employees]

    def take_attendance(self, name):
        return f"✓ Attendance taken for {name.title()}"

    def assign_task(self):
        task = random.choice(self.tasks)
        return f"➤ Assigned Task: {task.upper()}"

    def refuse_access(self):
        return "✗ ACCESS DENIED! You are not an employee."


# GUI Part
def process():
    name = entry.get()

    if emp.check_employee(name):
        result = emp.take_attendance(name) + "\n" + emp.assign_task()
    else:
        result = emp.refuse_access()

    output_label.config(text=result, fg="darkblue")


# Create object
emp = Employee()

# GUI window
root = tk.Tk()
root.title("Employee Attendance System")
root.geometry("450x350")
root.configure(bg="lightgray")

# Widgets
label = tk.Label(root, text="🔹 ENTER EMPLOYEE NAME:", font=("Arial", 10, "bold"), bg="lightgray")
label.pack(pady=10)

entry = tk.Entry(root, width=35, font=("Arial", 10))
entry.pack(pady=5)

button = tk.Button(root, text="SUBMIT", command=process, bg="navy", fg="white", font=("Arial", 9, "bold"))
button.pack(pady=10)

output_label = tk.Label(root, text="", wraplength=350, justify="center", font=("Arial", 9), bg="lightgray", fg="darkgreen")
output_label.pack(pady=20)

# Run GUI
root.mainloop()
```
